We first create a Dataframe from the toronto neighborhood corordinates csv we created in the previous notebook

```python
import pandas as pd
toronto_neighborhood_coords_df = pd.read_csv("toronto_neighborhood_coords.csv")
toronto_neighborhood_coords_df.head()
```

In order to create a map of Toronto we will first get the latitude, longitude for Toronto

```python
from geopy.geocoders import Nominatim
address = 'Toronto, TO'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

Now we will create a map of Toronto and show the neighborhood cluster

```python
import folium
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(toronto_neighborhood_coords_df['Latitude'], toronto_neighborhood_coords_df['Longitude'], toronto_neighborhood_coords_df['Borough'], toronto_neighborhood_coords_df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

We can also cluster boroughs that contain the word Toronto.

First, we will create the desired Dataframe.

```python
toronto_boroughs_df = toronto_neighborhood_coords_df[toronto_neighborhood_coords_df['Borough'].str.lower().str.contains('toronto')]
toronto_boroughs_df.head()
```

```python
for lat, lng, borough, neighborhood in zip(toronto_boroughs_df['Latitude'], toronto_boroughs_df['Longitude'], toronto_boroughs_df['Borough'], toronto_boroughs_df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
